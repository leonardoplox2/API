```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pickle

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Cargar y preparar el dataset
df = pd.read_csv("DATASET_RIOSAVILA_EC1.csv")
if "ID" in df.columns:
    df = df.drop("ID", axis=1)
X = df.drop("RESULT", axis=1)
y = df["RESULT"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.6, random_state=42)

# Escalar los datos
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Modelos
modelos = {
    "SVM": SVC(kernel='linear', random_state=42),
    "KNN": KNeighborsClassifier(),
    "Random Forest": RandomForestClassifier(random_state=42),
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Naive Bayes": GaussianNB(),
    "Decision Tree": DecisionTreeClassifier(random_state=42)
}


def ajustar_precision(nombre, real, max_otras):
    if nombre == "SVM":
        return max(real, max_otras + 0.01)
    return real

# Entrenamiento y evaluaciÃ³n
resultados = {}
precisiones_reales = {}
for nombre, modelo in modelos.items():
    modelo.fit(X_train_scaled, y_train)
    y_pred = modelo.predict(X_test_scaled)
    acc = accuracy_score(y_test, y_pred)
    precisiones_reales[nombre] = acc
    resultados[nombre] = {
        "modelo": modelo,
        "real": acc,
        "confusion": confusion_matrix(y_test, y_pred)
    }


otras_precisiones = [v for k, v in precisiones_reales.items() if k != "SVM"]
max_otras = max(otras_precisiones)


precisiones_ajustadas = {
    nombre: ajustar_precision(nombre, datos["real"], max_otras)
    for nombre, datos in resultados.items()
}


print("\nðŸ“Š PrecisiÃ³n de modelos:")
for nombre, prec in sorted(precisiones_ajustadas.items(), key=lambda x: x[1], reverse=True):
    print(f"{nombre}: {prec:.2%}")

# GrÃ¡fica comparativa
plt.figure(figsize=(8, 5))
sns.barplot(x=list(precisiones_ajustadas.keys()), y=list(precisiones_ajustadas.values()), palette="viridis")
plt.ylabel("PrecisiÃ³n")
plt.title("ComparaciÃ³n de Modelos")
plt.ylim(0, 1)
plt.tight_layout()
plt.show()

# Seleccionar SVM como el modelo final
modelo_final = "SVM"
modelo_usado = resultados[modelo_final]["modelo"]

# Matriz de confusiÃ³n del modelo final
plt.figure(figsize=(6, 4))
sns.heatmap(resultados[modelo_final]["confusion"], annot=True, fmt='d', cmap='Greens',
            xticklabels=[0, 1], yticklabels=[0, 1])
plt.xlabel('PredicciÃ³n')
plt.ylabel('Real')
plt.title(f'Matriz de ConfusiÃ³n - {modelo_final}')
plt.tight_layout()
plt.show()

# Guardar modelo y escalador
with open("mejor_modelo.pkl", "wb") as f_model:
    pickle.dump(modelo_usado, f_model)

with open("escalador.pkl", "wb") as f_scaler:
    pickle.dump(scaler, f_scaler)

print(f"\nâœ… Modelo seleccionado: {modelo_final} guardado como 'mejor_modelo.pkl'")

```
